```python
# Importar las libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from scipy.cluster.hierarchy import dendrogram, linkage

# Generar un conjunto de datos sinttico
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)

# Aplicar K-medias
kmeans = KMeans(n_clusters=4)
labels_kmeans = kmeans.fit_predict(X)

# Aplicar Agrupamiento Jerrquico
agglo = AgglomerativeClustering(n_clusters=4)
labels_agglo = agglo.fit_predict(X)

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Visualizar los resultados de K-medias
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.scatter(X[:, 0], X[:, 1], c=labels_kmeans, cmap='viridis', s=50)
plt.title("K-medias")

# Visualizar los resultados de Agrupamiento Jerrquico
plt.subplot(1, 3, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels_agglo, cmap='viridis', s=50)
plt.title("Agrupamiento Jerrquico")

# Visualizar los resultados de DBSCAN
plt.subplot(1, 3, 3)
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan, cmap='viridis', s=50)
plt.title("DBSCAN")
plt.show()
```

Anlisis Comparativo

K-medias: En este conjunto de datos, K-medias forma clusters esfricos de tamaos similares, lo que es consistente con la naturaleza de los datos. Sin embargo, la necesidad de predefinir el nmero de clusters y su sensibilidad a la inicializacin puede limitar su aplicabilidad en datos ms complejos.

Agrupamiento Jerrquico: El enfoque jerrquico proporciona una visin flexible de los datos, permitiendo la exploracin de diferentes niveles de agrupamiento. Sin embargo, la fusin irrevocable de clusters puede llevar a agrupaciones que no son ptimas si se selecciona un corte en el dendrograma que no refleje la estructura real de los datos.

DBSCAN: DBSCAN detecta clusters con formas ms arbitrarias y maneja eficientemente los outliers. Sin embargo, la sensibilidad a los parmetros `eps` y `min_samples` puede dificultar su aplicacin si no se seleccionan adecuadamente.
