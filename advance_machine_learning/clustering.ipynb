```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generar un dataset sintético con 300 muestras y 4 centros
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualización inicial de los datos
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Distribución Inicial de los Datos")
plt.show()
```

```python
# Aplicar k-medias con 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualizar los clusters resultantes
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red')
plt.title("Clusters Formados por k-medias")
plt.show()
```

Este ejercicio práctico con k-medias proporciona una primera experiencia en la creación y visualización de agrupamientos. Se puede observar cómo los datos se agrupan de manera natural en torno a los centros, y cómo k-medias es capaz de identificar estos grupos de manera efectiva. Esta técnica es una herramienta poderosa para la identificación de patrones en datos sin etiquetas y sienta las bases para exploraciones más avanzadas en técnicas de agrupamiento.
