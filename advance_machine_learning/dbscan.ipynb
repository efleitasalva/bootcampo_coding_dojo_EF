```python
# Importar las libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generar un conjunto de datos sinttico con forma de media luna
X, _ = make_moons(n_samples=300, noise=0.05)

# Aplicar DBSCAN con parmetros iniciales
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Resultados de DBSCAN con eps=0.2 y min_samples=5")
plt.show()
```

```python
# Aplicar DBSCAN con un valor de eps mayor
dbscan_eps_high = DBSCAN(eps=0.4, min_samples=5)
labels_eps_high = dbscan_eps_high.fit_predict(X)

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels_eps_high, cmap='viridis', s=50)
plt.title("Resultados de DBSCAN con eps=0.3 y min_samples=5")
plt.show()

# Aplicar DBSCAN con un valor de min_samples mayor
dbscan_min_samples_high = DBSCAN(eps=0.2, min_samples=10)
labels_min_samples_high = dbscan_min_samples_high.fit_predict(X)

# Visualizar los resultados
plt.scatter(X[:, 0], X[:, 1], c=labels_min_samples_high, cmap='viridis', s=50)
plt.title("Resultados de DBSCAN con eps=0.2 y min_samples=10")
plt.show()
```

```python
# Importar el conjunto de datos de iris
from sklearn.datasets import load_iris
iris = load_iris()
X_iris = iris.data

# Aplicar DBSCAN
dbscan_iris = DBSCAN(eps=0.5, min_samples=5)
labels_iris = dbscan_iris.fit_predict(X_iris)

# Visualizar los resultados en dos dimensiones
plt.scatter(X_iris[:, 0], X_iris[:, 1], c=labels_iris, cmap='viridis', s=50)
plt.title("DBSCAN en el Conjunto de Datos de Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

Al aplicar DBSCAN al conjunto de datos de iris, podemos ver cmo el algoritmo identifica clusters basados en las caractersticas de las flores, sin necesidad de especificar el nmero de clusters de antemano. Los puntos que no se agrupan en ningn cluster se clasifican como ruido, lo que puede ser til para detectar outliers en el conjunto de datos
