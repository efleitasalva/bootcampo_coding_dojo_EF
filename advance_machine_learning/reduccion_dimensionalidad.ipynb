```python
# Importar las libreras necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Cargar el conjunto de datos de iris
iris = load_iris()
X = iris.data
y = iris.target

# Aplicar PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print(X_pca.shape)
# Visualizacin de PCA
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50)
plt.title("Reduccin de Dimensionalidad usando PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()

# Aplicar t-SNE
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)
print(X_tsne.shape)
# Visualizacin de t-SNE
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=50)
plt.title("Reduccin de Dimensionalidad usando t-SNE")
plt.show()

# Aplicar LDA
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)
print(X_lda.shape)
# Visualizacin de LDA
plt.figure(figsize=(8, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis', s=50)
plt.title("Reduccin de Dimensionalidad usando LDA")
plt.show()


```

PCA: En la visualizacin generada por PCA, se puede observar cmo las dos primeras componentes principales capturan la mayor parte de la varianza en los datos. PCA es particularmente efectivo para capturar la estructura global de los datos, pero puede no ser tan detallado en la preservacin de relaciones locales como t-SNE.

t-SNE: t-SNE proporciona una visualizacin ms detallada, capturando mejor las relaciones locales entre los puntos. Esto es evidente en la clara separacin de las clases en el grfico, lo que hace que t-SNE sea ideal para la visualizacin de datos complejos.

LDA: LDA, al ser una tcnica supervisada, muestra una excelente separacin de las clases en el espacio de menor dimensin, lo que lo hace til para tareas de clasificacin. En este caso, la proyeccin de los datos mediante LDA maximiza la separacin entre las clases, lo que facilita la clasificacin posterior.
