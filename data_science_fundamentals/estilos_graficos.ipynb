```python
import matplotlib.pyplot as plt
import numpy as np

# Aplicar un estilo predefinido
plt.style.use('ggplot')

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico
plt.plot(x, y)
plt.title('Gráfico de Seno con Estilo ggplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un gráfico con marcadores
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Gráfico de Seno con Marcadores')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Crear un gráfico con un título
plt.plot(x, y)
plt.title('Gráfico de Seno', fontsize=20, color='r', loc='center')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Crear un gráfico con etiquetas en los ejes
plt.plot(x, y)
plt.xlabel('Eje X', fontsize=15)
plt.ylabel('Eje Y', fontsize=15)
plt.title('Gráfico de Seno con Etiquetas')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Aplicar un estilo predefinido
plt.style.use('seaborn-v0_8-darkgrid')

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura con subplots
fig, ax = plt.subplots(figsize=(12, 6))

# Gráfico de Seno
ax.plot(x, y1, label='Seno', color='b', marker='o', linestyle='-', markersize=5)

# Gráfico de Coseno
ax.plot(x, y2, label='Coseno', color='r', marker='^', linestyle='--', markersize=5)

# Añadir título y etiquetas
ax.set_title('Gráficos de Seno y Coseno', fontsize=20)
ax.set_xlabel('Eje X', fontsize=15)
ax.set_ylabel('Eje Y', fontsize=15)

# Añadir leyenda
ax.legend(loc='upper right')

# Mostrar el gráfico
plt.show()
```
