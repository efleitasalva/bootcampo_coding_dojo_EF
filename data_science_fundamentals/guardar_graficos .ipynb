```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una nueva figura
fig = plt.figure(figsize=(10, 6))

# Aadir un subplot a la figura
ax = fig.add_subplot(111)  # 111 significa 1x1 cuadrcula, primer subplot

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear un grfico en el subplot
ax.plot(x, y, label='Seno')
ax.set_title('Grfico de Seno')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()

# Mostrar la figura
plt.show()
```

```python
# Guardar la figura en formato PNG
fig.savefig('grafico_seno.png')
```

```python
# Guardar la figura en formato PDF
fig.savefig('grafico_seno.pdf')
```

```python
# Guardar la figura en formato SVG
fig.savefig('grafico_seno.svg')
```

```python
# Guardar la figura con alta resolucin y fondo transparente
fig.savefig('grafico_seno_alta_resolucion.png', dpi=300, bbox_inches='tight', transparent=True)
```

Buenas Prcticas para Guardar Figuras

* Especificar la Resolucin: Utiliza el parmetro dpi para asegurarte de que la calidad de la imagen es adecuada para el uso previsto.
* Utilizar Fondos Transparentes: Especialmente til para presentaciones donde se requiere que la imagen se integre con diferentes fondos.
* Ajustar el Recorte: Utiliza bbox_inches=tight para eliminar espacios en blanco innecesarios alrededor de la figura.
* Nombrado Descriptivo de Archivos: Usa nombres de archivos que describan claramente el contenido del grfico.
