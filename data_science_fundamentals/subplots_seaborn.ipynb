```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una figura con 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Primer subplot
ax1.plot(x, y1, label='Seno')
ax1.set_title('Gráfico de Seno')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()

# Segundo subplot
ax2.plot(x, y2, label='Coseno', color='r')
ax2.set_title('Gráfico de Coseno')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.legend()

# Mostrar la figura con los subplots
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Subplot 1
axs[0, 0].plot(x, y1, 'tab:blue')
axs[0, 0].set_title('Seno')

# Subplot 2
axs[0, 1].plot(x, y2, 'tab:orange')
axs[0, 1].set_title('Coseno')

# Subplot 3
axs[1, 0].plot(x, -y1, 'tab:green')
axs[1, 0].set_title('Seno Negativo')

# Subplot 4
axs[1, 1].plot(x, -y2, 'tab:red')
axs[1, 1].set_title('Coseno Negativo')

for ax in axs.flat:
    ax.set(xlabel='X', ylabel='Y')

# Ajustar la posición de los subplots para evitar superposición
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Datos de ejemplo
df = sns.load_dataset('tips')

# Crear un gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# Crear un gráfico de barras
sns.barplot(data=df, x='day', y='total_bill', hue='sex')
plt.title('Total Bill por Día y Sexo')
plt.show()
```

```python
# Crear un gráfico de cajas
sns.boxplot(data=df, x='day', y='total_bill', hue='smoker')
plt.title('Distribución de Total Bill por Día y Fumadores')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', ax=axs[0, 0])
axs[0, 0].set_title('Total Bill vs Tip')

# Gráfico de barras
sns.barplot(data=df, x='day', y='total_bill', hue='sex', ax=axs[0, 1])
axs[0, 1].set_title('Total Bill por Día y Sexo')

# Gráfico de cajas
sns.boxplot(data=df, x='day', y='total_bill', hue='smoker', ax=axs[1, 0])
axs[1, 0].set_title('Distribución de Total Bill por Día y Fumadores')

# Histograma
sns.histplot(data=df, x='total_bill', kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Histograma de Total Bill')

# Ajustar la posición de los subplots para evitar superposición
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
# Generar datos falsos
np.random.seed(0)
size = 100
data = {
    'feature1': np.random.randn(size),
    'feature2': np.random.randn(size) * 2,
    'category': np.random.choice(['A', 'B', 'C'], size=size),
    'value': np.random.rand(size) * 100
}

df_fake = pd.DataFrame(data)
df_fake
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Scatter plot
sns.scatterplot(data=df_fake, x='feature1', y='value', hue='category', ax=axs[0, 0])
axs[0, 0].set_title('Feature 1 vs Value')

# Bar plot
sns.barplot(data=df_fake, x='category', y='value', ax=axs[0, 1])
axs[0, 1].set_title('Value by Category')

# Box plot
sns.boxplot(data=df_fake, x='category', y='feature2', ax=axs[1, 0])
axs[1, 0].set_title('Feature 2 by Category')

# Hist plot
sns.histplot(data=df_fake, x='value', bins=10, kde=True, ax=axs[1, 1])
axs[1, 1].set_title('Distribution of Value')

# Ajustar la posición de los subplots para evitar superposición
plt.tight_layout()
plt.show()
```
