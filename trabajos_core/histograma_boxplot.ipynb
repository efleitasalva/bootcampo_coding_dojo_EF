Cargar y Explorar los Datos:

* Descarga el dataset y cárgalo en un DataFrame de Pandas.
* Realiza una exploración inicial de los datos para entender las variables disponibles.



```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

### Para el caso de estudio, tomare el ultimo año que es el archivo 2019

```python

def cargar_datos(ruta_archivo):
    # Carga los datos del archivo CSV
    datos = pd.read_csv(ruta_archivo, delimiter=',')
    return datos
##Se realiza la carga con Pandas ya que numpy no muestra los valores categoricos
if __name__ == "__main__":
    ruta_archivo = '../trabajos_core/data/happines_2019.csv'
    df = cargar_datos(ruta_archivo) 

df.shape
df
```

```python
##informacion dataset
df.info()
##estadisticas descriptivas
df.describe()
##Detallado de valores nulos por columna
# Esto permite determinar cuántos valores en cada columna tienen datos cargados
qsna = df.shape[0] - df.isnull().sum(axis=0)

# Calcula la cantidad de valores nulos en cada columna del DataFrame
qna = df.isna().sum(axis=0)

# Calcula el porcentaje de valores nulos en cada columna y redondea a 2 decimales
ppna = round(100 * (df.isnull().sum(axis=0) / df.shape[0]), 2)

# Crea un diccionario auxiliar con la cantidad de datos sin nulos, cantidad de nulos y porcentaje de nulos
aux = {'datos sin NAs en q': qsna, 'Na en q': qna, 'Na en %': ppna}

# Convierte el diccionario en un DataFrame
na = pd.DataFrame(data=aux)

# Ordena el DataFrame por el porcentaje de valores nulos en orden descendente
na.sort_values(by='Na en %', ascending=False)
```

##### No se encuentran valores nulos  y los tipos de datos son dtypes: float64(7 columns), int64(1 column), object(1 column)

Análisis Combinado:

* Combina los histogramas y los boxplots en una sola figura para cada variable.
* Interpreta los resultados y escribe un análisis sobre la distribución de cada variable y la presencia de valores atípicos.



```python
# Crear una figura con subplots
fig, ax = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 4]})

# Crear el boxplot en el subplot superior
sns.boxplot(x=df['GDP per capita'], ax=ax[0], color='lightblue')
ax[0].set_title('Boxplot de GDP per capita')

# Crear el histograma en el subplot inferior
sns.histplot(df['GDP per capita'], bins=10, kde=False, ax=ax[1], color='blue')
ax[1].set_title('Histograma de GDP per capita')

# Ajustar el layout para evitar solapamiento
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

##### El boxplot nos muestra que para la variable GDP per capita la media se encuentra al rededor de 0.90, El rango intercuartílico (IQR), que representa el rango entre el primer cuartil (Q1) y el tercer cuartil (Q3), muestra que la mayor parte de los datos está entre aproximadamente 0.75 y 1.24. No parece haber valores atípicos (outliers) en este conjunto, ya que no se observan puntos fuera de los bigotes. Los bigotes muestran el rango total de los datos, que va desde alrededor de 0.25 hasta 1.5.

##### En el caso del Histograma, los datos de GDP per capita están algo distribuidos de forma asimétrica hacia la derecha, ya que hay más países con valores más bajos entre 0.5 y 1.0 que con valores mayores a 1.25.La mayor concentración de países está en el rango entre 0.75 y 1.25, también hay algunos valores cercanos a 0.

```python
# Crear una figura con subplots
fig, ax = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 4]})

# Crear el boxplot en el subplot superior
sns.boxplot(x=df['Score'], ax=ax[0], color='lightblue')
ax[0].set_title('Boxplot de Score')

# Crear el histograma en el subplot inferior
sns.histplot(df['Score'], bins=10, kde=False, ax=ax[1], color='blue')
ax[1].set_title('Histograma de Score')

# Ajustar el layout para evitar solapamiento
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

##### El boxplot muestra que la mediana del "Score" es de aproximadamente 5.5, los cuartiles Q1 y Q3 son aproximadamente de 4.5 y 6.5. No se observan valores atipicos fuera del rango esperado.

##### En el historgrama se peude ver que la distribucion es aproximadamente normal, tambien se puede ver que no hay barras aisladas muy alejadas del resto lo que nos muestra una ausencia de valores atipicos. 

```python
# Crear una figura con subplots
fig, ax = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 4]})

# Crear el boxplot en el subplot superior
sns.boxplot(x=df['Healthy life expectancy'], ax=ax[0], color='lightblue')
ax[0].set_title('Boxplot de Healthy life expectancy')

# Crear el histograma en el subplot inferior
sns.histplot(df['Healthy life expectancy'], bins=10, kde=False, ax=ax[1], color='blue')
ax[1].set_title('Histograma de Healthy life expectancy')

# Ajustar el layout para evitar solapamiento
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

##### En el boxplot de esta variable se puede observar que existe un valor atipico mayor de esperanza de vida saludable, lo cual es una outlier que podria ser atipico o especial que merece mas estudio. La mediana se encuentra aproximadamente en 0.7 y los valores IQR Q1 y Q3 se encuentran entre 0.55 y 0.9 respectivamente. Se muestra una ligera asimetria a la izquierda, lo que indica que hay mas paises con una esperanza de vida saludable por encima de la mediana.

##### Para el caso del histograma, la forma del historgrama confirma una asimetria a la izquierda como el boxplot que sugiere que la mayoria de los paises "felices" tienen alto indice una espectativa de vida alta.

```python

# Crear una figura con subplots
fig, ax = plt.subplots(2, 1, figsize=(8, 6), gridspec_kw={'height_ratios': [1, 4]})

# Crear el boxplot en el subplot superior
sns.boxplot(x=df['Freedom to make life choices'], ax=ax[0], color='lightblue')
ax[0].set_title('Boxplot de Freedom to make life choices')

# Crear el histograma en el subplot inferior
sns.histplot(df['Freedom to make life choices'], bins=10, kde=False, ax=ax[1], color='blue')
ax[1].set_title('Histograma de Freedom to make life choices')

# Ajustar el layout para evitar solapamiento
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

##### En el boxplot podemos observar un punto fuera de la caja, lo que representa que hay un valor atipico que seria un pais o region donde la libertad de tomar decisiones de vida es significativamente menor al resto. La caja que representa el IRQ se muestra en un tamano moderado lo que suguiere variabilidad en los datos. La mediana se encuentra mas para arriba de la caja entre el 0.4 y 0.5 lo indica que la mayoria de los paises tienen un nivel relativamente alto de esta variable.

##### Para el caso del histograma, la asimetria a la derecha confirma que la mayoria de paises para la libertad de tomar decisiones de vida se encuentra en el 0.4 y 0.5 y luego van disminuyendo. Se puede observar que en el nivel 0.4 hay una tendencia que se considera la moda por se la mayoria.
